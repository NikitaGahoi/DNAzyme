```python
browser = Browser('chrome')
```

```python
# Visit the website
# https://www.genesilico.pl/DNAmoreDB/dnazymes
main_url = "https://www.genesilico.pl/DNAmoreDB/dnazymes"
browser.visit(main_url)
```

```python
# Create a Beautiful Soup object
html = browser.html
zyme_soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data cities = soup.find('ol', id='cities')
table= zyme_soup.find('table', id ='dnazymes_table')
```

```python
header = table.find_all('th')
```

```python
type(header)
```

```python
#extract all the headers
title =[]
for heading in header:
    value = heading.text.strip()
    title.append(value)
    #to remove blanks
    rm_blank = filter(None, title)
    f_title = list(rm_blank)
print(f_title)
```

```python
#creating dataframe using ordered dictionary method- Method 1
#Extracting all the values and creating an ordered dictionary to create a dataframe with the extracted headers and values
Zyme_temp = []
od_Zyme_temp = []

for row in rows:
    columns = row.find_all('td')
    if len(columns)>0:
        name = columns[0].get_text()
        zyme_link = columns[0].a['href']
        length = columns[1].get_text()
        reaction  = columns[3].get_text()
        cofactor = columns[4].get_text()
        #print(name, Length, zyme_link, Reaction, cofactor)
        
        # Automate clicking the next button
        browser.links.find_by_partial_text(name.strip()).click()
        html = browser.html
        seq_soup = soup(html, 'html.parser')
        sequence_element = seq_soup.find('td', class_='sequence')
        sequence = sequence_element.text.strip()
        
        
        browser.links.find_by_partial_text("Go Back to Dnazymes").click()
        html = browser.html
        zyme_soup = soup(html, 'html.parser')
        
        Zyme_temp.append([name, length, sequence, reaction, cofactor])
        od = co.OrderedDict(zip(f_title,([name, length, sequence, reaction, cofactor])))
        od_Zyme_temp.append(od)
```

```python
DNAzyme_DF = pd.DataFrame(od_Zyme_temp)
DNAzyme_DF
```

```python
# Write the data to a CSV
DNAzyme_DF.to_csv("DNA_zyme_final.csv", header=True, index=False)
```
